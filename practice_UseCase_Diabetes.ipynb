```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diabetesData=pd.read_csv('diabetes.csv')
```

```python
diabetesData_copy=diabetesData.copy(deep=True)
```

```python
diabetesData_copy
```

```python
diabetesData_copy.isnull()
```

```python
diabetesData_copy.isnull().sum()
```

```python
diabetesData_copy.columns
```

```python
diabetesData_copy[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI']]=diabetesData_copy[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI']].replace(0,np.NaN)
```

```python
diabetesData_copy.isnull().sum()
```

```python
p=diabetesData_copy.hist(figsize=(20,20))
```

```python
diabetesData_copy.boxplot(figsize=(40,20),fontsize=18)
```

```python
diabetesData_copy['Glucose'].fillna(diabetesData_copy['Glucose'].mean(),inplace=True)
```

```python
diabetesData_copy.isnull().sum()
```

```python
diabetesData_copy['BloodPressure'].fillna(diabetesData_copy['BloodPressure'].mean(),inplace=True)
```

```python
diabetesData_copy.isnull().sum()
```

```python
diabetesData_copy['SkinThickness'].fillna(diabetesData_copy['SkinThickness'].median(),inplace=True)
```

```python
diabetesData_copy.isnull().sum()
```

```python
diabetesData_copy['Insulin'].fillna(diabetesData_copy['Insulin'].median(),inplace=True)
```

```python
diabetesData_copy.isnull().sum()
```

```python
diabetesData_copy['BMI'].fillna(diabetesData_copy['BMI'].median(),inplace=True)
```

```python
diabetesData_copy.isnull().sum()
```
